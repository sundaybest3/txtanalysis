```python
from matplotlib import pyplot as plt
import numpy as np

# Data (percentage values)
functions = ["DE", "CON", "COR", "TM", "DIS", "TRM"]
textbook = [61.94, 8.39, 2.58, 10.97, 16.13, 0.00]
sbcsae = [26.83, 4.39, 8.29, 45.86, 3.41, 11.22]

# Bar positions
x = np.arange(len(functions))
width = 0.35

# Create grayscale-friendly clustered bar chart
plt.figure(figsize=(10,6))

plt.bar(
    x - width/2,
    textbook,
    width,
    label="Textbook",
    edgecolor='black',
    hatch='///'
)

plt.bar(
    x + width/2,
    sbcsae,
    width,
    label="SBCSAE",
    edgecolor='black',
    hatch='...'
)

# Axis labels
plt.xticks(x, functions)
plt.ylabel("Percentage (%)")

# Legend (APA-friendly minimal frame)
plt.legend(frameon=False)

# Save figure (optional)
plt.savefig("but_function_clustered_bar_chart_bw.png", bbox_inches='tight')

# Show figure
plt.show()
```
