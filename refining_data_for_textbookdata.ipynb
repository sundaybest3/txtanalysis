```
import pandas as pd
import re
from google.colab import files

# =========================
# 1. GitHub CSV 불러오기
# =========================
github_url = "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/rawdata/High%20School%20Textbook%20Dialogue%20data_first%20analysis.xlsx.csv"
csv_url = github_url.replace("github.com", "raw.githubusercontent.com").replace("/blob/", "/")

df = pd.read_csv(csv_url)

print(df.columns.tolist())  # 열 이름 확인

# =========================
# 2. 정제 함수
# =========================

def clean_sbcsae_for_wordcount(text):
    if pd.isna(text):
        return text

    text = str(text)

    # --------------------------------
    # A. 단어 내부 기호 먼저 제거 (붙이기)
    # --------------------------------

    text = re.sub(r':+', '', text)      # A:nd → And
    text = re.sub(r'ʔ', '', text)
    text = re.sub(r'%', '', text)

    # --------------------------------
    # B. SBCSAE annotation 제거 (공백 유지)
    # --------------------------------

    text = re.sub(r'&\{[^}]*\}', ' ', text)
    text = re.sub(r'&\}[^ ]*', ' ', text)

    text = re.sub(r'&=\S+', ' ', text)
    text = re.sub(r'=\S+', ' ', text)

    text = re.sub(r'\(\.{1,}\)', ' ', text)
    text = re.sub(r'\+\S*', ' ', text)

    text = re.sub(r'\bX+\b', ' ', text)
    text = re.sub(r'\b\d+\b', ' ', text)

    # overlap markers
    text = re.sub(r'[⌈⌉⌊⌋]', '', text)

    # --------------------------------
    # C. 일반 문장부호 제거 (붙이기)
    # --------------------------------

    text = re.sub(r"[.,!?;:\"()[\]{}<>]", '', text)

    # --------------------------------
    # D. 영어 단어/공백/' 만 남기기
    # --------------------------------

    # Added '’' (curly apostrophe) to the allowed characters to ensure it is kept.
    text = re.sub(r"[^A-Za-z'’\s]", ' ', text)

    # --------------------------------
    # E. 공백 정리
    # --------------------------------

    text = re.sub(r'\s+', ' ', text).strip()

    return text

    # =========================
# 3. script 열에 덮어쓰기
# =========================
df["Script"] = df["Script"].apply(clean_sbcsae_for_wordcount)

# =========================
# 4. 결과 일부 확인
# =========================
print(df["Script"].head(20))

# =========================
# 5. 저장
# =========================
output_file = "cleaned_textbook_data.csv"
df.to_csv(output_file, index=False, encoding="utf-8-sig")

files.download(output_file)
```
