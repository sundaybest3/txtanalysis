```python
import pandas as pd
import re
from google.colab import files

# =========================
# 1. GitHub CSV 불러오기
# =========================
github_url = "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/main/sbcase_56.csv"
csv_url = github_url.replace("github.com", "raw.githubusercontent.com").replace("/blob/", "/")

df = pd.read_csv(csv_url)

print(df.columns.tolist())  # 열 이름 확인

# =========================
# 2. 정제 함수
# =========================
def clean_sbcsae_for_wordcount(text):
    if pd.isna(text):
        return text

    text = str(text)

    # --------------------------------
    # A. 먼저 특정 SBCSAE 표지 제거
    # --------------------------------

    # 범위 태그 제거: &{l=MRC ... &}l=MRC
    text = re.sub(r'&\{[^}]*\}', ' ', text)
    text = re.sub(r'&\}[^ ]*', ' ', text)

    # &=something 제거
    text = re.sub(r'&=\S+', ' ', text)

    # =something 제거  ← 이번에 강하게 처리
    text = re.sub(r'=\S+', ' ', text)

    # pause 표기 제거: (.), (..), (...)
    text = re.sub(r'\(\.{1,}\)', ' ', text)

    # cut-off 표기 제거: +/., +//., +/, +// 등
    text = re.sub(r'\+\S*', ' ', text)

    # XX, XXX 등 제거
    text = re.sub(r'\bX+\b', ' ', text)

    # 단독 숫자 제거
    text = re.sub(r'\b\d+\b', ' ', text)

    # --------------------------------
    # B. 자주 나오는 특수기호 제거
    # --------------------------------
    text = re.sub(r'[⌈⌉⌊⌋]', '', text)   # overlap
    text = re.sub(r'ʔ', '', text)        # glottal stop
    text = re.sub(r'%', '', text)
    text = re.sub(r':+', '', text)
    text = re.sub(r'&', ' ', text)

    # --------------------------------
    # C. 영어 단어/공백/아포스트로피만 남기기
    #    word count 목적이면 이 단계가 핵심
    # --------------------------------
    text = re.sub(r"[^A-Za-z'\s]", ' ', text)

    # --------------------------------
    # D. 공백 정리
    # --------------------------------
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# =========================
# 3. script 열에 덮어쓰기
# =========================
df["script"] = df["script"].apply(clean_sbcsae_for_wordcount)

# =========================
# 4. 결과 일부 확인
# =========================
print(df["script"].head(20))

# =========================
# 5. 저장
# =========================
output_file = "cleaned_sbcase_56.csv"
df.to_csv(output_file, index=False, encoding="utf-8-sig")

files.download(output_file)
```
