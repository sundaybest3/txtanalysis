```python
from matplotlib import pyplot as plt
import numpy as np

# Data (percentage values)
functions = ["DE", "CON", "COR", "TM", "DIS", "TRM"]
textbook = [61.94, 8.39, 2.58, 10.97, 16.13, 0.00]
sbcsae = [26.83, 4.39, 8.29, 45.86, 3.41, 11.22]

# Bar positions
x = np.arange(len(functions))
width = 0.35

# Create grayscale-friendly clustered bar chart
plt.figure(figsize=(14,10)) # Increased figure size for better readability

# Set a larger base font size for better readability
plt.rcParams.update({'font.size': 18})

plt.bar(
    x - width/2,
    textbook,
    width,
    label="Textbook",
    edgecolor='black',
    hatch='///'
)

plt.bar(
    x + width/2,
    sbcsae,
    width,
    label="SBCSAE",
    edgecolor='black',
    hatch='...'
)

# Axis labels with increased font size
plt.xticks(x, functions, fontsize=20)
plt.ylabel("Percentage (%)", fontsize=22)
plt.xlabel("Functions", fontsize=22)

# Legend (APA-friendly minimal frame) with increased font size
plt.legend(frameon=False, fontsize=20)

# Save figure (optional)
plt.savefig("but_function_clustered_bar_chart_bw.png", bbox_inches='tight')

# Show figure
plt.show()
```

```python
# Percentage data
functions = ["DE", "CON", "COR", "TM", "DIS", "TRM"]

textbook = [61.94, 8.39, 2.58, 10.97, 16.13, 0.00]
sbcsae   = [26.83, 4.39, 8.29, 45.85, 3.41, 11.22]

# Create figure
plt.figure(figsize=(8,6))

# Draw slope lines
for i in range(len(functions)):
    plt.plot(
        [0, 1],
        [textbook[i], sbcsae[i]],
        marker='o',
        linewidth=1.5,
        color='black'
    )

# X-axis labels
plt.xticks([0, 1], ["Textbook", "SBCSAE"])

# Y-axis label
plt.ylabel("Percentage (%)")

# Function labels at midpoints
for i, func in enumerate(functions):
    mid_y = (textbook[i] + sbcsae[i]) / 2
    plt.text(
        0.5,
        mid_y,
        func,
        ha='center',
        va='center',
        fontsize=10
    )

# Remove top/right borders (journal style)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Save figure
plt.savefig("but_function_slope_chart_bw.png", bbox_inches='tight')

# Show figure
plt.show()
```

```python
import numpy as np

# Data (percentages)
functions = ["DE", "CON", "COR", "TM", "DIS", "TRM"]

textbook = np.array([61.94, 8.39, 2.58, 10.97, 16.13, 0.00])
sbcsae   = np.array([26.83, 4.39, 8.29, 45.85, 3.41, 11.22])

# Combine data
data = np.vstack([textbook, sbcsae])

# Plot
fig, ax = plt.subplots(figsize=(8,6))

bottom = np.zeros(2)

# grayscale-friendly hatch patterns
hatches = ['///', '...', '\\\\', 'xxx', '---', '+++']

for i, func in enumerate(functions):
    values = data[:, i]

    ax.bar(
        ["Textbook", "SBCSAE"],
        values,
        bottom=bottom,
        label=func,
        edgecolor='black',
        hatch=hatches[i]
    )

    bottom += values

# Axis label
ax.set_ylabel("Percentage (%)")

# Legend outside (journal-friendly)
ax.legend(
    title="Function",
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.tight_layout()

# Save figure
plt.savefig("but_function_stacked_bar_chart_bw.png")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

functions = ["DE", "TM", "TRM", "DIS", "COR", "CON"]

textbook = np.array([61.94, 10.97, 0.00, 16.13, 2.58, 8.39])
sbcsae   = np.array([26.83, 45.85, 11.22, 3.41, 8.29, 4.39])

data = np.vstack([textbook, sbcsae])

fig, ax = plt.subplots(figsize=(5.2, 2.6))

left = np.zeros(2)

hatches = ['///', 'xxx', '+++', '---', '\\\\', '...']

for i, func in enumerate(functions):
    values = data[:, i]
    ax.barh(
        ["Textbook", "SBCSAE"],
        values,
        left=left,
        label=func,
        edgecolor="black",
        hatch=hatches[i],
        height=0.6
    )
    left += values

ax.set_xlabel("Percentage (%)", fontsize=10)
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)

ax.invert_yaxis()

ax.legend(
    title="Function",
    bbox_to_anchor=(1.05, 1),
    loc="upper left",
    frameon=False,
    fontsize=10,
    title_fontsize=11
)

plt.tight_layout()
plt.savefig("but_function_stacked_bar_chart_horizontal.png", bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# standardized residuals 값 입력
residuals = np.array([
    [3.84, 1.15, -1.68, -4.45, 3.02, -3.15],
    [-3.34, -1.00, 1.46, 3.87, -2.63, 2.74]
])

# dataframe 변환
df_residuals = pd.DataFrame(
    residuals,
    index=["Textbook", "SBCSAE"],
    columns=["DE", "CON", "COR", "TM", "DIS", "TRM"]
)

# heatmap 그리기
plt.figure(figsize=(5.2, 2.4))

ax = sns.heatmap(
    df_residuals,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    annot_kws={"size": 11},
    cbar_kws={"label": "Standardized Residuals"}
)

# |residual| ≥ 2 이면 흰색 글씨 처리
for text in ax.texts:
    value = float(text.get_text())
    if abs(value) >= 2:
        text.set_color("white")
    else:
        text.set_color("black")

# 축 글씨 크기 조정
plt.xticks(fontsize=11, rotation=0)
plt.yticks(fontsize=11)

# colorbar 글씨 크기 조정
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=10)
cbar.set_label("Standardized Residuals", fontsize=11)

plt.tight_layout()

# 저장
plt.savefig("standardized_residuals_heatmap.png", dpi=300, bbox_inches="tight")

plt.show()
```
