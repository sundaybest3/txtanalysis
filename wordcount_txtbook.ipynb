```
# 1. 라이브러리 불러오기
import pandas as pd

# 2. GitHub raw CSV 링크
url = "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/main/High%20School%20Textbook%20Dialogue%20data_first%20analysis.xlsx%20-%20sheet1.csv"

# 3. CSV 불러오기
df = pd.read_csv(url)

# 4. 열 이름 확인 (선택)
print(df.columns)

# 5. word count 할 열 이름
column_name = "Script"

# 6. 해당 열 전체 word count
text = " ".join(df[column_name].dropna())
word_count = len(text.split())

print("Total word count:", word_count)
```
