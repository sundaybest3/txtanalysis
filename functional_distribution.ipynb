```python
import pandas as pd

# 1️⃣ 데이터 불러오기
url = "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/refineddata/High%20School%20Textbook%20Dialogue%20data_first%20analysis%20(1).csv"
df = pd.read_csv(url, encoding='euc-kr')

# 2️⃣ 열 이름 확인 (Function 열 이름 정확히 확인용)
print(df.columns)


# 3️⃣ 필요한 열 선택 (예: Label, Function_A)
# 'Function' 열이 없으므로 'Function_A'를 사용하고, crosstab을 위해 필요한 열만 선택합니다.
df = df[['Label', 'Agreement']].dropna()


# 4️⃣ 교과서 × 기능별 빈도표 생성
table = pd.crosstab(df['Label'], df['Agreement'])


# 5️⃣ 행별 total 추가
table['Total'] = table.sum(axis=1)


# 6️⃣ 열별 total 추가
total_row = table.sum(axis=0)
total_row.name = 'Total'

table = pd.concat([table, total_row.to_frame().T])


# 7️⃣ 결과 출력
print(table)
```

```python
import pandas as pd

# Define table as it was in cell xNg2IQLiiTfK to ensure it's available
url = "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/refineddata/High%20School%20Textbook%20Dialogue%20data_first%20analysis%20(1).csv"
df = pd.read_csv(url, encoding='euc-kr')
df = df[['Label', 'Agreement']].dropna()
table = pd.crosstab(df['Label'], df['Agreement'])
table['Total'] = table.sum(axis=1)
total_row = table.sum(axis=0)
total_row.name = 'Total'
table = pd.concat([table, total_row.to_frame().T])

# 9️⃣ 각 행별 퍼센테이지 계산
# 'Total' 열을 제외하고 퍼센테이지를 계산합니다.
percentage_table = table.iloc[:-1, :-1].div(table.iloc[:-1, -1], axis=0) * 100

# 'Total' 행과 'Total' 열은 퍼센테이지 계산에서 제외하거나, 적절히 처리합니다.
# 여기서는 'Total' 행을 포함한 전체 테이블의 각 행의 Total 값을 기준으로 퍼센테이지를 계산합니다.
# 단, 최종 'Total' 행은 그대로 유지합니다.
percentage_table_with_total = table.iloc[:, :-1].div(table['Total'], axis=0) * 100

# 마지막 'Total' 행은 퍼센테이지 계산에서 제외하고 빈도수를 유지하거나, 전체에 대한 퍼센티지로 계산할 수 있습니다.
# 여기서는 각 'Label' (교과서) 내에서 'Agreement' 분포의 퍼센테이지를 보여주는 것이 목적이므로,
# 각 Label의 Total을 기준으로 퍼센티지를 계산하고, 최종 Total 행은 합계 퍼센티지로 보여줍니다.

# 각 'Label' 행의 퍼센테이지 계산 (마지막 'Total' 행 제외)
percentage_table = table.iloc[:-1, :-1].div(table.iloc[:-1, -1], axis=0) * 100

# 'Total' 열 추가 (원본 Total 값 사용)
percentage_table['Total'] = table.iloc[:-1]['Total']

# 마지막 'Total' 행 추가 (원본 Total 값 사용)
# 'Total' 행의 퍼센테이지는 전체 빈도 대비 각 기능의 퍼센티지로 계산할 수 있으나,
# 여기서는 각 교과서별 분포를 보는 것이 주 목적이므로, 원본의 'Total' 행을 그대로 붙여넣습니다.
# 또는 전체에 대한 퍼센티지를 계산할 수도 있습니다.
# 예시: Total 행의 각 기능별 퍼센티지 (전체 Total을 100%로)
overall_total = table.loc['Total', 'Total']
overall_percentage_row = (table.loc['Total', :].iloc[:-1] / overall_total) * 100
overall_percentage_row['Total'] = 100.0 # 전체 Total은 100%
overall_percentage_row.name = 'Total'

percentage_table = pd.concat([percentage_table, overall_percentage_row.to_frame().T])


# 10️⃣ 결과 출력 (소수점 두 자리까지)
print("\n교과서 × 기능별 빈도 퍼센테이지표:")
print(percentage_table.round(2))
```

```python
# 8️⃣ 결과 테이블을 CSV 파일로 저장
output_filename = 'frequency_table.csv'
table.to_csv(output_filename, encoding='euc-kr')
print(f"'{output_filename}' 파일로 결과가 저장되었습니다.")
```
