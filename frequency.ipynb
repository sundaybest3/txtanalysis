```
import pandas as pd
import re

# 1️⃣ GitHub CSV 파일 불러오기
url = "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/rawdata/High%20School%20Textbook%20Dialogue%20data_first%20analysis.xlsx.csv"
df = pd.read_csv(url)

# 2️⃣ 분석할 열 지정 (예: script 열)
column_name = "Script"

# 결측값 제거
texts = df[column_name].dropna().astype(str)

# 3️⃣ 전체 텍스트 합치기
full_text = " ".join(texts)

# 4️⃣ 전체 단어 수 계산
words = re.findall(r'\b\w+\b', full_text.lower())
total_word_count = len(words)

# 5️⃣ but 등장 횟수 계산 (대소문자 구분 없음)
but_count = len(re.findall(r'\bbut\b', full_text.lower()))

# 6️⃣ per 10,000 words 기준 빈도 계산
normalized_frequency = (but_count / total_word_count) * 10000

# 7️⃣ 결과 출력
print("Total word count:", total_word_count)
print("Frequency of 'but':", but_count)
print("Frequency per 10,000 words:", round(normalized_frequency, 1))
```

```
import re

# 'but'이(가) 나타나는 모든 경우를 찾아서 주변 텍스트와 함께 출력합니다.
# 대소문자를 구분하지 않기 위해 full_text를 소문자로 변환합니다.
search_text = full_text.lower()

# 'but'이(가) 나타나는 모든 위치를 찾습니다.
# 정규 표현식 r'\bbut\b'는 독립적인 단어 'but'을 찾습니다.
matches = list(re.finditer(r'\bbut\b', search_text))

print(f"'but' 단어의 총 {len(matches)}개의 발생을 찾았습니다:\n")

# 각 'but'의 발생과 그 주변 텍스트를 출력합니다.
context_length = 50  # 'but' 단어의 앞뒤로 50글자의 맥락을 보여줍니다.

for i, match in enumerate(matches):
    start_index = match.start()
    end_index = match.end()

    # 'but' 단어 앞의 맥락을 가져옵니다.
    start_context = max(0, start_index - context_length)
    before_but = full_text[start_context:start_index]

    # 'but' 단어 뒤의 맥락을 가져옵니다.
    end_context = min(len(full_text), end_index + context_length)
    after_but = full_text[end_index:end_context]

    # 'but' 단어 자체는 원본 텍스트에서 가져옵니다 (대소문자 유지).
    original_but = full_text[start_index:end_index]

    print(f"[{i+1}] ...{before_but}**{original_but}**{after_but}...")
    print("--------------------------------------------------")
```

```
import pandas as pd

# 결과를 저장할 리스트 초기화
but_occurrences_data = []

# 각 'but'의 발생과 그 주변 텍스트를 추출하여 리스트에 추가합니다.
for i, match in enumerate(matches):
    start_index = match.start()
    end_index = match.end()

    # 'but' 단어 앞의 맥락을 가져옵니다.
    start_context = max(0, start_index - context_length)
    before_but = full_text[start_context:start_index]

    # 'but' 단어 뒤의 맥락을 가져옵니다.
    end_context = min(len(full_text), end_index + context_length)
    after_but = full_text[end_index:end_context]

    # 'but' 단어 자체는 원본 텍스트에서 가져옵니다 (대소문자 유지).
    original_but = full_text[start_index:end_index]

    but_occurrences_data.append({
        'occurrence_number': i + 1,
        'before_but': before_but,
        'but_word': original_but,
        'after_but': after_but,
        'full_context': f"...{before_but}**{original_but}**{after_but}..."
    })

# 리스트를 데이터프레임으로 변환
df_but_occurrences = pd.DataFrame(but_occurrences_data)

# CSV 파일로 저장
output_filename = 'but_occurrences.csv'
df_but_occurrences.to_csv(output_filename, index=False, encoding='utf-8')

print(f"'{output_filename}' 파일에 'but' 발생 정보가 저장되었습니다.")

# 저장된 데이터프레임의 첫 5행을 출력하여 확인
display(df_but_occurrences.head())
```

```
import pandas as pd
import re

# 1️⃣ 데이터 불러오기
url = "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/rawdata/High%20School%20Textbook%20Dialogue%20data_first%20analysis_2.xlsx.csv"
df = pd.read_csv(url)

# 2️⃣ 열 이름 정리
df.columns = df.columns.str.strip()

# 3️⃣ 필요한 열 선택
df = df[['Label', 'Script']].copy()

# 4️⃣ 결측값 처리
df['Script'] = df['Script'].fillna('').astype(str)

# 5️⃣ apostrophe 두 종류 모두 포함한 word count 계산
df['word_count'] = df['Script'].apply(
    lambda x: len(re.findall(r"[A-Za-z]+(?:['’][A-Za-z]+)?", x))
)

# 6️⃣ but 등장 횟수 계산
df['but_frequency'] = df['Script'].str.lower().str.count(r'\bbut\b')

# 7️⃣ 교과서별 집계
result = df.groupby('Label', as_index=False).agg(
    total_words=('word_count', 'sum'),
    but_frequency=('but_frequency', 'sum')
)

# 8️⃣ per 10,000 words 계산
result['per_10000_words'] = (
    result['but_frequency'] /
    result['total_words'] * 10000
).round(1)

# 9️⃣ 정렬
result = result.sort_values('Label')

# 🔟 전체 합계 계산
total_words_sum = result['total_words'].sum()
but_frequency_sum = result['but_frequency'].sum()

total_row = pd.DataFrame({
    'Label': ['TOTAL'],
    'total_words': [total_words_sum],
    'but_frequency': [but_frequency_sum],
    'per_10000_words': [
        round((but_frequency_sum / total_words_sum) * 10000, 1)
    ]
})

# 11️⃣ 합계 행 추가
result = pd.concat([result, total_row], ignore_index=True)

# 12️⃣ 결과 출력
print(result)
```

```
import pandas as pd
import re

url = "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/rawdata/High%20School%20Textbook%20Dialogue%20data_first%20analysis_2.xlsx.csv"
df = pd.read_csv(url)

df.columns = df.columns.str.strip()
df = df[['Label', 'Script']].copy()
df['Script'] = df['Script'].fillna('').astype(str)

# 하이픈과 apostrophe를 단어 내부 요소로 허용
def count_words(text):
    tokens = re.findall(r"[A-Za-z]+(?:[-’'][A-Za-z]+)*", text)
    return len(tokens)

df['word_count'] = df['Script'].apply(count_words)
df['but_frequency'] = df['Script'].str.lower().str.count(r'\bbut\b')

result = df.groupby('Label', as_index=False).agg(
    total_words=('word_count', 'sum'),
    but_frequency=('but_frequency', 'sum')
)

result['per_10000_words'] = (
    result['but_frequency'] / result['total_words'] * 10000
).round(1)

result = result.sort_values('Label')

total_row = pd.DataFrame({
    'Label': ['TOTAL'],
    'total_words': [result['total_words'].sum()],
    'but_frequency': [result['but_frequency'].sum()],
    'per_10000_words': [round(result['but_frequency'].sum() / result['total_words'].sum() * 10000, 1)]
})

result = pd.concat([result, total_row], ignore_index=True)

print(result)
```
