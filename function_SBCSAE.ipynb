```python
import pandas as pd

# 1️⃣ CSV URL 목록 입력
urls = [
    "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/sbcsae_final/cleaned_sbcase_47.csv",
    "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/sbcsae_final/cleaned_sbcase_49.csv",
    "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/sbcsae_final/cleaned_sbcase_50.csv",
    "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/sbcsae_final/cleaned_sbcase_51.csv",
    "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/sbcsae_final/cleaned_sbcase_56.csv"
]

# 2️⃣ 모든 파일 읽기
dfs = [pd.read_csv(url) for url in urls]

# 3️⃣ 하나의 데이터프레임으로 합치기
df_all = pd.concat(dfs, ignore_index=True)

# 4️⃣ Function 열 선택 (열 이름 확인 필요)
print(df_all.columns)

# 예: 'Function' 열 기준 집계
function_counts = df_all['agreement'].value_counts().sort_index()

# 5️⃣ 결과 출력
print(function_counts)

# 6️⃣ 전체 합계 계산 및 출력
total_agreement_count = function_counts.sum()
print(f"\nTotal agreement count: {total_agreement_count}")

# 7️⃣ 각 agreement 유형의 백분율 계산 및 출력 (소수점 셋째 자리에서 반올림)
agreement_percentages = (function_counts / total_agreement_count * 100).round(2)
print("\nAgreement Percentages:")
print(agreement_percentages)
```
