```python
import numpy as np
from scipy.stats import chi2_contingency

# Observed frequencies
# 순서: DE, CON, COR, TM, DIS, TRM
textbook = [96, 13, 4, 17, 25, 0]
sbcsae = [55, 9, 17, 94, 7, 23]

# Contingency table
observed = np.array([textbook, sbcsae])

# Chi-square test
chi2, p, dof, expected = chi2_contingency(observed)

print("Chi-square:", chi2)
print("Degrees of freedom:", dof)
print("p-value:", p)
print("\nExpected frequencies:\n", expected)
```

```python

n = observed.sum()
cramers_v = np.sqrt(chi2 / (n * (min(observed.shape)-1)))

print("Cramér's V:", cramers_v)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Observed frequencies
functions = ["DE", "CON", "COR", "TM", "DIS", "TRM"]
textbook = [96, 13, 4, 17, 25, 0]
sbcsae = [55, 9, 17, 94, 7, 23]

total_textbook = sum(textbook)
total_sbcsae = sum(sbcsae)

results = []

for func, t, s in zip(functions, textbook, sbcsae):
    observed = np.array([
        [t, total_textbook - t],
        [s, total_sbcsae - s]
    ])

    ll, p, dof, expected = chi2_contingency(observed, lambda_="log-likelihood")

    results.append({
        "Function": func,
        "LL": round(ll, 2),
        "p-value": p
    })

df = pd.DataFrame(results)
print(df)
```
