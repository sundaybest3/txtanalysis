```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score

# 1️⃣ 여러 데이터 URL 목록
urls = [
    "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/refineddata/High%20School%20Textbook%20Dialogue%20data_first%20analysis.xlsx.csv",
    "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/refineddata/cleaned_sbcase_47.csv",
    "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/refineddata/cleaned_sbcase_49.csv",
    "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/refineddata/cleaned_sbcase_50.csv",
    "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/refineddata/cleaned_sbcase_51.csv",
    "https://raw.githubusercontent.com/sundaybest3/txtanalysis/refs/heads/refineddata/cleaned_sbcase_56.csv"
]

# 2️⃣ 모든 데이터 불러오기
dfs = [pd.read_csv(url) for url in urls]

# 3️⃣ 하나로 합치기
df_all = pd.concat(dfs, ignore_index=True)

# 4️⃣ 결측값 제거
df_all = df_all.dropna(subset=['function1', 'function2'])

# 5️⃣ percent agreement 계산
agreement = (df_all['function1'] == df_all['function2']).mean()

# 6️⃣ Cohen's kappa 계산
kappa = cohen_kappa_score(df_all['function1'], df_all['function2'])

# 7️⃣ 결과 출력
print("Total instances:", len(df_all))
print("Percent agreement:", round(agreement * 100, 2), "%")
print("Cohen's kappa:", round(kappa, 3))
```
