```python
import pandas as pd
from scipy.stats import spearmanr
from google.colab import files

# 1️⃣ CSV 파일 직접 사용 (frequency_by_label.csv가 이미 업로드되어 있다고 가정)
filename = '/content/sample_data/frequency_by_label.csv'

# 2️⃣ CSV 읽기
df = pd.read_csv(filename)

# 3️⃣ 열 이름 확인
print("Columns:", df.columns)

# 예: total_words, but_frequency 열이 있다고 가정
# 필요하면 아래 이름 수정 가능

# 4️⃣ TOTAL 행이 있으면 제거
df = df[df['Label'] != 'TOTAL']

# 5️⃣ Spearman correlation 계산
# 현재 코드에서는 'total_words'와 'but_frequency' 열을 사용하고 있습니다.
# 파일에 해당 열이 있는지 확인하거나, 필요한 열 이름으로 수정해주세요.
rho, p_value = spearmanr(df['total_words'], df['but_frequency'])

# 6️⃣ 결과 출력
print(f"Spearman's rho = {rho:.3f}")
print(f"p-value = {p_value:.4f}")

if p_value < 0.05:
    print("There is a statistically significant correlation.")
else:
    print("There is no statistically significant correlation.")
```
